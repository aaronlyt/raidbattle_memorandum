# 補遺　タンパク質編\_第０章　タンパク質に関する情報の収集



本編ではスクリーニングの処理手順記録をまとめました。in silicoでスクリーニングを行うにあたって、**標的のタンパク質**に関する情報と**既知の活性リガンド**に関する情報はとても有用な情報となります。リガンドについては大変ありがたいことに「**共闘用シェアデータ**」という形で情報を利用することができました。また、タンパク質側についても複数の**共結晶構造**が報告されており構造情報を入手可能でした。

タンパク質構造情報を利用するにあたって公共データベースの利用や、その質の検証などを行いました。折角ですので「**補遺　タンパク質編**」として、こちらの内容についてもまとめたいと思います。スクリーニング編同様ブログ記事の再掲載という形ですがご了承ください。

内容は以下のようになっています。

第１章ではまず、「どのような構造情報を手に入れることができるのか？」、複数のデータベースの情報が統合されている**UniProt**を利用した構造情報の一覧等の取得を試みています。ついで、結晶構造データベースである**PDB**を利用し、同ページで提供されている　ビューワー上で共結晶構造の確認を行いました（第２章）。

第３章では「リガンド結合ポケットとして狙うべき場所が共結晶構造の結合部位で良いのか？」、**CASTp**を使ってありえる結合ポケットの位置を探索しています。スクリーニング編におけるSBVSとも関わってくる話題かと思います。

また、第４章から第７章では、PDBに登録されている「共結晶構造中のリガンド構造が間違っているのではないか？」という疑問から出発し、その理由を探るため

* PDBファイルに記載されている情報の確認（フォーマットの見方）（第４章）
* 結晶学の非専門家がPDBの情報を使う上で注意したいこと（第５章）
* PDBのビューワーを使って情報の質を検証する方法（第６章）
* フリーソフフトウェアのcootを使って、電子密度と構造の比較を行う手順　（第７章）

といったことを試みています。

記載している本人が結晶学の専門家ではなく、門外漢のため調べながら記載した内容となっています。初歩的な内容で、正しさを保証するものではありませんが、ご参考までにしていただければ幸いです。
