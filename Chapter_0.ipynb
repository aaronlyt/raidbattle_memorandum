# 第０章　はじめに


このページはソーシャルメディア上で創薬に関する情報発信、企画を実施されている @souyakuchan 主催の 「誰でも参加可能な薬剤探索イベント　**創薬レイドバトル**」に参加し、*in silico* screening の実施を試みた記録を残すために作成しました。イベントの詳細な説明については下記ページをご参照ください。

[創薬レイドバトル 2018: 腫瘍免疫回避機構 討伐戦](https://gist.github.com/souyakuchan/7f9add656d9d6b91f02f6939b88aa2c8#創薬レイドバトル-2018-腫瘍免疫回避機構-討伐戦)

簡単に概要をまとめると以下となります。  
* 目的) PD/PD-L1 免疫チェックポイント阻害剤に利用可能な低分子化合物の探索
* 方法) 参加者各自で候補化合物を絞り込み運営に提出  
* 検証) 運営で実際のアッセイを行い阻害活性を確認

また、留意点として  
「参加者同士で腕を競い勝敗を決めるためのフェアなコンテストではなく、あくまで**創薬**することを目的としたイベント（腫瘍免疫回避機構をみんなの力で打倒するレイドバトル）」  
と、されています。

上記イベントに参加するにあたって、*Virtual Screening*、* Chemoinfomatics* という分野を１から勉強することになったのですが、幸いにも様々な先生方からご支援、ご指導をいただくことができました。ありがとうございました。

折角教えていただいたことを忘れてしまわないためにも、行った処理の記録を備忘録としてこちらのページでまとめなおすこととしました。間違いだらけの酷い記録ですが、初心者はこういうところにつまづくのか、とご笑覧いただけたら幸いです。
